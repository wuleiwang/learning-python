本笔记使用ipython notebook工具，使用技巧包括：
1. 使用例如？min然后点击运行的方式可以查看某个特定函数的意义和用法；
2. 使用%matlotlib inline让运行的结果图嵌入到浏览器中；
3. 代码自动补全功能，比如输入该关键词的一到两个字母，后面按tab可以自动补齐；
4. 按enter+shift可以自动运行当前所在cell，并另新开一个cell。

```python
# 注意如果要图像显示在浏览器中应该
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

t=np.arange(0,1,0.01)

# 画出红色的线条图形
s=np.sin(2*np.pi*t)
plt.rcParams['lines.color']='r'
plt.plot(t,s)

# 设定线条的粗细
c=np.cos(2*np.pi*t)
plt.rcParams['lines.linewidth']='1'
plt.plot(t,c)

plt.show()
plt.rcdefaults()
```
